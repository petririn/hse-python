```python
# задача 0
numbers = []

while True:
    n = int(input())
    if n == 0:
        break
    numbers += [n] if n > 0 and n % 2 == 0 else []

print(numbers, len(numbers),  sep='\n')
```

```python
#  задача 1
celsius = [0, 15, 25, 30, 40, 100]
fahrenheit = [c * 9/5 + 32 for c in celsius]

print(f"Цельсий:{celsius}\nФаренгейт:{fahrenheit}")

grades = [45, 85, 92, 33, 67, 78, 90, 55, 29, 88]
passed = [g for g in grades if g >= 60]

print(f'''
Все оценки: {grades}
Прошли: {passed}
Количество сдавших: {len(passed)}
''')
```

```python
# задача 2
transactions = [100, -50, 200, -30, 150, -20, 300]

taxes = [t * 0.15 for t in transactions if t > 0]

print(f'''
Транзакции: {transactions}
Налоги: {taxes}
''')
```

```python
# задача 3
products = ['футболка', 'кружка', 'блокнот']
colors = ['красный', 'синий', 'зеленый']

combinations = [f"{product} - {color}" for product in products for color in colors]

for combo in combinations:
  print(combo)
```

```python
# задача 4
employees = [('Иван', 45), ('Мария', 92), ('Петр', 33), ('Анна', 67)]

for name, score in employees:
    perfomance = ("Отлично" if score >= 90 else
                  "Хорошо" if score >= 60 else
                  "Требует улучшения")
    print(f"{name} ({score}%): {perfomance}")
```

```python
# задача 5

sales = [
    {'name': 'Алексей', 'sales': 150, 'returns': 10},
    {'name': 'Ольга', 'sales': 200, 'returns': 5},
    {'name': 'Дмитрий', 'sales': 80, 'returns': 25},
    {'name': 'Елена', 'sales': 300, 'returns': 8}
]

top_managers = [
    sale['name']
    for sale in sales
    if (sale['sales'] - sale['returns']) >= 150 and (sale['returns'] / sale['sales']) * 100 < 10
]

print(top_managers)
```
